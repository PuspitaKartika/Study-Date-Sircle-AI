```python
import matplotlib.pyplot as plt
```

```python
x = [0,10]
y = [100,250]

plt.plot(x,y,)
plt.xlabel("Baris X")
plt.ylabel("Baris Y")
plt.title("Study date SIRCLE") #judul

plt.grid()
plt.show()
```

```python
import numpy as np

angka = np.array([3,7,2,8])
plt.plot(angka)
plt.show()
```

SCATTER PLOT

```python
x = np.array([5,6,3,7,1])
y = np.array([22,34,23,12,20])

plt.rcParams['axes.facecolor'] = "white"
plt.scatter(x,y, color = "red")
plt.show()
```

BAR

```python
x = np.array(["A","B","C","D"])
y = np.array([3,6,2,4])

plt.bar(x,y, color = "red", width =0.5)
plt.show()
```

PIE CHAR

```python
y = np.array([3,5,8,1])
buah = np.array(["jeruk","semangka","anggur","mangga"])

plt.pie(y, labels = buah)
plt.legend()
plt.show()
```
